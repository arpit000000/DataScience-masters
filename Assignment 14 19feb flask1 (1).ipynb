Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python. It is designed to be lightweight and simple, allowing developers to build web applications quickly and with minimal code. Flask follows the Model-View-Controller (MVC) architectural pattern and is often used to create RESTful APIs, web services, and small to medium-sized web applications.

Here are some advantages of using Flask:

Lightweight and flexible: Flask is a minimalistic framework that provides just the essentials for web development. It allows developers to choose the components they need, making it flexible and suitable for various projects.

Easy to get started: Flask has a straightforward and intuitive API, making it easy for developers, especially beginners, to understand and start building applications quickly. It has a small learning curve compared to other frameworks.

Modular and extensible: Flask follows a modular design, which means it can be easily extended with additional functionalities. It provides a wide range of extensions to add features such as database integration, authentication, and testing.

Built-in development server: Flask comes with a built-in development server, allowing developers to test and run their applications without the need for additional setup. It simplifies the development process and makes it convenient for local development.

Jinja2 templating: Flask uses the Jinja2 templating engine, which provides powerful and flexible templates for rendering dynamic content. It separates the presentation logic from the business logic, resulting in more maintainable code.

Large and active community: Flask has a large and active community of developers, which means there are plenty of resources, tutorials, and extensions available. This makes it easier to find solutions to problems and get support when needed.

Well-documented: Flask has excellent documentation, which is crucial for any framework or library. The documentation covers all the essential aspects of Flask, including detailed examples and explanations, making it easier to learn and use.

Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to specific functions or views within your application. It allows you to define the endpoints or routes that your application will respond to when receiving HTTP requests.

App routes are used to define the different pages or actions that your web application can perform. Each route corresponds to a URL that users can visit or interact with. When a user visits a specific URL in the browser or sends a request to that URL, Flask matches the URL to the corresponding route and executes the associated function or view.
App routes are essential because they define the structure and functionality of your web application. They allow you to handle different URLs and perform specific actions or display different content based on the user's request. By defining routes, you can create a navigation structure and enable users to interact with different pages or perform various operations within your application.
