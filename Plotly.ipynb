```python
import plotly.graph_objects as go
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=[1,2,3,4,5],y=[9,8,10,7,11],mode='markers'))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=[1,2,3,4,5],y=[9,8,10,7,11]))
fig.show()
```

```python
x=[1,2,3,2,1,3,5,4,8,6,9,8,1,2,3,1,2,3,4,3,2,1]
fig=go.Figure(data=[go.Histogram(x=x)])
fig.show()
```

```python
tips=sns.load_dataset('tips')
fig=go.Figure(data=[go.Histogram(x=tips.total_bill)])
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=tips.total_bill,y=tips.tip,mode='markers',marker_size=10*tips['size']))
fig.show()
```

```python
#3D
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,mode='markers',z=tips['size']))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter3d(x=tips.total_bill,y=tips.tip,mode='lines',z=tips['size']))
fig.show()
```

```python
#Bokeh library
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
output_file('test.html')
p=figure(title='test flowers')
p.xaxis.axis_label='petal_length'
p.yaxis.axis_label='petal_width'
p.circle(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
from bokeh.plotting import figure,output_file,show
from bokeh.sampledata.iris import flowers
output_file('test.html')
p=figure(title='line grapg')
p.xaxis.axis_label='petal_length'
p.yaxis.axis_label='petal_width'
p.line(flowers['petal_length'],flowers['petal_width'])
show(p)
```

```python
x=[1,2,3,4,5]
y=[9,8,7,6,5]
output_file('test.html')
p=figure(title='plot')
p.scatter(x,y,size=20,legend='red points',fill_color='red')
show(p)
```
