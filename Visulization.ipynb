```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(1,16,200)
y=np.sin(x)
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.title('Data')
plt.xlabel('Values')
plt.ylabel('Sin')
plt.grid()
plt.figure(figsize=(3,3))
plt.show()
```

```python
plt.plot(x,y,'--r')
plt.title('Data')
plt.xlabel('Values')
plt.ylabel('Sin')
plt.show()
```

```python
import math
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y,c='Red',s=120,alpha=.3,marker='x')#s represent size and alpha represent intensity of colour marker give your symbol
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Data')
plt.show()
```

```python
#categorical vs numerical
x=['a','b','c','d','e']
y=np.random.randint(1,15,size=(5))
plt.bar(x,y)
plt.xlabel('Name')
plt.ylabel('Sales')
plt.title('BAR GRAPH')
plt.show()
```

```python
#histogram represent data repetation
data=[1,2,3,1,2,3,4,3,2,1,5,7,6,9]
plt.hist(data)
plt.show()
```

```python
#colour=['FF5733','Blue','Yellow','20C1C1','CA3AEE']
x=[1,2,3,4,5]
y=[12,53,86,35,44]
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
fig.add_subplot(projection='3d')
plt.scatter(x,y,z)
plt.show()
```
